```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
```

```python
train=pd.read_csv("train.csv")
train
```

```python
test=pd.read_csv("test_share.csv")
test
```

```python
medical=pd.read_csv("medical_history.csv")
medical
```

```python
demographic=pd.read_csv("demographic_details.csv")
demographic 
```

```python
# Merge datasets
train_data = pd.merge(train, medical, on='PatientId', how='left')
train_data = pd.merge(train, demographic, on='PatientId', how='left')
train_data
```

```python
train_data.isnull().sum()
```

```python
#'No-show' is target variable
y = train_data['No-show']

# Convert 'Yes' and 'No' to binary values
y = y.map({'Yes': 1, 'No': 0})
```

```python
# 'No-show' is target variable
X = train_data.drop(columns=['No-show'])
```

```python
# Print unique values in the 'ScheduledDay' and 'AppointmentDay' columns after conversion
print("Unique values in 'ScheduledDay' column:", X['ScheduledDay'].unique())
print("Unique values in 'AppointmentDay' column:", X['AppointmentDay'].unique())
```

```python
# Convert date columns to datetime objects
X['ScheduledDay'] = pd.to_datetime(X['ScheduledDay'])
X['AppointmentDay'] = pd.to_datetime(X['AppointmentDay'])
```

```python
# Extract features from datetime columns
X['ScheduledDay_year'] = X['ScheduledDay'].dt.year
X['ScheduledDay_month'] = X['ScheduledDay'].dt.month
X['ScheduledDay_day'] = X['ScheduledDay'].dt.day
X['ScheduledDay_hour'] = X['ScheduledDay'].dt.hour
```

```python
# Drop unnecessary columns
X = X.drop(columns=['ScheduledDay', 'AppointmentDay'])
```

```python
# One-hot encode categorical columns with handling unseen values
X_encoded = pd.get_dummies(X, columns=['Gender', 'Neighbourhood'], dummy_na=True)
```

```python
# Identify and scale numeric columns
numeric_columns = X_encoded.select_dtypes(include=['number']).columns
scaler = StandardScaler()
X_scaled_numeric = scaler.fit_transform(X_encoded[numeric_columns])
```

```python
# Concatenate scaled numeric features with one-hot encoded categorical features
X_scaled = pd.concat([pd.DataFrame(X_scaled_numeric, columns=numeric_columns), X_encoded.drop(columns=numeric_columns)], axis=1)
```

```python
# Split the data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size=0.1, random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
```

```python
# Create and train a Logistic Regression model
model = LogisticRegression(random_state=10)
model.fit(X_train, y_train)
```

```python
# Predict probabilities for the validation set
y_val_pred_prob = model.predict_proba(X_val)[:, 1]
```

```python
# Evaluate the model using ROC-AUC score
roc_auc = roc_auc_score(y_val, y_val_pred_prob)
print(f'ROC-AUC Score on Validation Set: {roc_auc}')
```
